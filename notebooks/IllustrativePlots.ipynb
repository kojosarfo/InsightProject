Here are some illustrative plots of pedestrians/ cyclist traffic in Downtown Toronto

```python
from __future__ import print_function
import pickle
#import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading saved files

filename = 'PedestriansData.sav'
Pedestrians = pickle.load(open(filename, 'rb'))

filename = 'CyclistsData.sav'
Cyclists = pickle.load(open(filename, 'rb'))
```

Example pedestrian traffic in a day for a given intersection. Here, the traffic for 31 August is shown; may have to change date to date-of-interest

```python
#Example pedestrian traffic in a day
Peds = Pedestrians[0] #The index here ranges from 0 to 13, and refer to different intersections

startTime = pd.to_datetime('2019-08-30 00:00:00');
endTime = pd.to_datetime('2019-08-30 23:00:00');

IntersectionName = Peds['IntName'][0]

X = Peds[startTime: endTime]
X = X.dropna()
y = X['volume']
x = y.index
plt.plot(x.hour, y, 'r')
plt.grid(True); plt.title('Pedestrian traffic at ' + IntersectionName + ' intersection')
plt.xlabel('30 August 2019'); plt.ylabel('Pedestrian volume')
plt.show()
```

Example cyclist traffic in a day. Here, the traffic for 31 August is shown; may have to change date to date-of-interest.

```python
#Example cyclist traffic in a day
Cycs = Cyclists[0]; #The index here ranges from 0 to 13, and refer to different intersections

startTime = pd.to_datetime('2019-08-30 00:00:00');
endTime = pd.to_datetime('2019-08-30 23:00:00');

IntersectionName = Cycs['IntName'][0]

X = Cycs[startTime: endTime]
X = X.dropna()
y = X['volume']
x = y.index
plt.plot(x.hour, y, 'r')
plt.grid(True); plt.title('Cyclist traffic at ' + IntersectionName + ' intersection')
plt.xlabel('30 August 2019'); plt.ylabel('Cyclist volume')
plt.show()
```

Illustration of missing data: gap-filling

```python
#Intersection number 9
Cycs = Cyclists[9]

startTime = pd.to_datetime('2019-05-01 00:00:00');
endTime = pd.to_datetime('2019-05-31 23:00:00');

IntersectionName = Cycs['IntName'][0]

X = Cycs[startTime: endTime]
X = X.dropna()
y = X['volume']
x = y.index
plt.plot(x.day, y, 'ro')
plt.grid(True); plt.title('Cyclist traffic at ' + IntersectionName + ' intersection')
plt.xlabel('May 2019'); plt.ylabel('Cyclist volume')
plt.show()
```

Pedestrian traffic distribution at different hours for different intersections

```python
for i in range(14):
    
    Peds = Pedestrians[i]
    Peds = Peds.dropna()
    IntersectionName = Peds['IntName'][0]
    
    dailyData = [];
    for j in range(24):
        jthHourData = Peds[Peds['Hour'] == j]['volume']
        dailyData.append(jthHourData.values)
    boxplot = plt.boxplot(dailyData)
    plt.xlabel('Time (hours)'); plt.ylabel('Pedestrian traffic volume')
    plt.title('Pedestrian traffic at ' + IntersectionName + ' intersection')
    plt.grid(True)
    plt.show()
```

The cell below shows the distribution of traffic volume at the (n+1)th hour

```python
n = 16;
sns.kdeplot(dailyData[n]);
plt.grid(True)
plt.title('Pedestrian traffic distribution at '+ str(n+1)+ ': 00')
```
